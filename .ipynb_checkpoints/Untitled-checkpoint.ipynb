```python
import pandas as pd
```

```python
df_temp = pd.read_csv("global_temp.csv")
df_temp
```

```python
df_temp['dt'] = pd.to_datetime(df_temp['Year'],format='%Y')
df_temp.index = pd.DatetimeIndex(df_temp['dt']).year
# df_temp.index = pd.to_datetime(df_temp.dt, format='%Y')
df_temp = df_temp.drop(columns = ['Year','dt'])
```

```python
df_temp
```

```python
training_data = df_temp[df_temp.index < 2015]
```

```python
training_data
```

```python
test_data = df_temp[-5:]
```

```python
len(test_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
diff = np.diff(df_temp.Temperature)
```

```python
diff
```

```python
plt.plot(diff)
```

```python
from statsmodels.tsa.stattools import adfuller

X = diff
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(training_data,lags = 10)
plot_pacf(training_data,lags = 10,method='ywm')
plt.show()
```

```python
import traceback
import logging
from matplotlib.pyplot import figure
from statsmodels.tsa.arima.model import ARIMA

p_range = list(range(0,5))  # taking values from 0 to 2
q_range = list(range(0,5))
aic_values = []
bic_values = []
pq_values = []
fitted_model_dict = {}
count = 0
d =2
for p in p_range:
    for q in q_range:
        try:
            count +=1
            model = ARIMA(training_data, order=(p, d, q))
            results = model.fit()
            aic_values.append(results.aic)
            bic_values.append(results.bic)
            pq_values.append((p, q))
        except Exception as e:
            pass
#             logging.error(traceback.format_exc())


best_pq = pq_values[aic_values.index(min(aic_values))]  # (p,q) corresponding to lowest AIC score
print("(p,q) corresponding to lowest AIC score: ", best_pq)
```

```python
results.plot_diagnostics(figsize=(15, 18))
plt.show()
```

```python
arima_model = ARIMA(training_data, order=(best_pq[0], 0, best_pq[1])).fit()
predictions = arima_model.predict(start=166, end=len(df_temp)-1)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(list(test_data['Temperature']), list(predictions))
print("Mean Squared Error:", mse)

mae = mean_absolute_error(list(test_data['Temperature']), list(predictions))
print("Mean Absolute Error:", mae)
```
